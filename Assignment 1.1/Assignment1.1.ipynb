```python
import pickle
import numpy as np

with open("/home/saurabhmishra/Classroom/GenAI/Datasets/Embeddings saved as pickle/saved_embeddings.pickle", "rb") as p:
    embed=pickle.load(p)
```

```python
from sentence_transformers import util

embed_len=len(embed.keys())
matrix=[[0]*embed_len for i in range(embed_len)]
# matrix=np.zeros((embed_len,embed_len))
for i, text_i in enumerate(embed.keys()):
    for j, text_j in enumerate(embed.keys()):
        matrix[i][j]=np.round(util.cos_sim(embed[text_i],embed[text_j])[0][0],5)
```

```python
import seaborn as sns
import matplotlib as plt
ax = sns.heatmap(matrix,annot=True,xticklabels=embed.keys(),yticklabels=embed.keys())
```
